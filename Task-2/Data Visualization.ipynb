```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_data = pd.read_csv(url)
```

```python
# 1. Data Cleaning 
# Drop rows with missing Age
titanic_data = titanic_data.dropna(subset=["Age"])
```

```python
# 2. Visualizations

# a. Distribution of age by survival status
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_data, x="Age", hue="Survived", multiple="stack", kde=True, bins=30)
plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(["Not Survived", "Survived"])
plt.show()
```

```python
# b. Bar plot of survival rate by class 
plt.figure(figsize=(8, 5))
sns.barplot(data=titanic_data, x="Pclass", y="Survived", errorbar=None, palette="muted")
plt.title("Survival Rate by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# c. Boxplot of fare distribution across different classes
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_data, x="Pclass", y="Fare", palette="coolwarm")
plt.title("Fare Distribution Across Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
# d. Pie chart of survival proportion
survival_counts = titanic_data["Survived"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=["Not Survived", "Survived"], autopct="%1.1f%%", startangle=140, colors=["red", "green"])
plt.title("Survival Proportion")
plt.show()
```

```python
# e. Correlation heatmap of numerical variables (updated)
plt.figure(figsize=(10, 8))
correlation_matrix = titanic_data.corr(numeric_only=True)  # Explicitly include only numeric columns
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
